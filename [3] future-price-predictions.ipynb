```python
import yfinance as yf
import pandas as pd
import numpy as np
import cv2
import os
import torch 
from ultralytics import YOLO
import glob
from IPython.display import display, HTML, Image 
import base64 
import warnings 
import random

# --- WARNING SUPPRESSION (unchanged) ---
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
# --- END WARNING SUPPRESSION ---


# --- Utility Functions (unchanged) ---

def error_line(message):
    """Prints a formatted error message."""
    print('âŒâŒâŒâŒâŒâŒ')
    print(message)

def create_folder(path):
    """Creates a directory if it doesn't exist."""
    try:
        os.makedirs(path, exist_ok=True)
        print("ğŸ“ Folder created or already exists: " + path) # ğŸ“ log message
    except Exception as e:
        error_line(f"Error creating folder {path}: {e}")
        
def render_result(model, image, result):
    """Draws YOLO results on the image and returns a numpy array."""
    # ğŸ–¼ï¸ This uses the built-in ultralytics plotting function.
    return result.plot()

def load_model(model_path):
    """Loads the YOLO model using the PyTorch environment variable bypass."""
    try:
        print('Loading model ...')
        
        os.environ['TORCH_FORCE_NO_WEIGHTS_ONLY_LOAD'] = '1'
        print("ğŸ”§ Set TORCH_FORCE_NO_WEIGHTS_ONLY_LOAD=1 for security bypass.") # ğŸ”§ log message
        
        # The UserWarning is suppressed globally now, making this call silent.
        model = YOLO(model_path)
        
        # Clean up the environment variable after use
        del os.environ['TORCH_FORCE_NO_WEIGHTS_ONLY_LOAD']
        
        print("âœ… Model loaded successfully (Environment Variable Bypass).") # âœ… log message
        
        # âš™ï¸ set model parameters (These are base parameters, the trading threshold is set in draw_data)
        model.overrides['conf'] = 0.25 
        model.overrides['iou'] = 0.45
        model.overrides['agnostic_nms'] = False
        model.overrides['max_det'] = 5
        
        return model
    except Exception as e:
        error_line(f"Final error loading model: {e}")
        error_line("Failed to load model. Please check the 'best.pt' file integrity.")
        return None

# --- backtest function (unchanged) ---
def backtest(predictions, entry_prices, actual_outcome):
    """
    Backtests the predictions and calculates win rate and total cumulative return.
    Returns: [win_rate, total_trades, cumulative_return_factor]
    """
    try:
        total = len(predictions)
        win_count = 0
        cumulative_return_factor = 1.0 # Initial capital is 1.0 (100%)

        # ğŸ”„ Comparing prediction to actual outcome and compounding returns.
        for prediction, entry_price, the_actual_outcome in zip(predictions, entry_prices, actual_outcome):
            
            is_win = False
            raw_return_factor = 1.0 # Default to no change if not predicted

            if prediction == "up":
                # Long trade: return = Outcome / Entry
                raw_return_factor = the_actual_outcome / entry_price
                if the_actual_outcome > entry_price:
                    is_win = True
                    
            elif prediction == "down":
                # Short trade: return = 1 + (Entry/Outcome - 1)
                raw_return_factor = entry_price / the_actual_outcome
                if the_actual_outcome < entry_price:
                    is_win = True
                
            if is_win:
                win_count += 1
                
            # ğŸ“ˆ Compounding the return factor
            cumulative_return_factor *= raw_return_factor
        
        # Win rate is still useful
        win_rate = (win_count / total) * 100 if total > 0 else 0
        
        return [win_rate, total, cumulative_return_factor] 
        
    except Exception as e:
        error_line(f"Backtest error: {e}")
        return [0, 0, 1.0]
```

```python
# --- Main Class (TradingBot) ---

class TradingBot:
    """A trading class that handles data download, chart generation, prediction, and video output."""
    
    # ğŸ› ï¸ Initialized with flexible bars and threshold parameters
    def __init__(self, model_path: str, bars: int = 40, threshold: float = 0.6):
        self.model = load_model(model_path)
        if self.model is None:
            raise Exception("Model could not be loaded. Cannot proceed.")
        self.bars = bars 
        self.threshold = threshold 
        print(f"âš™ï¸ Bot initialized with {self.bars} bars and a trade confidence threshold of {self.threshold}.") # âš™ï¸ log message

    # --- RESTORED: download_data method (CORRECTLY INDENTED) ---
    def download_data(self, ticker: str, start_date: str, end_date: str, timeframe: str = '1d') -> pd.DataFrame or None:
        """Downloads stock data and cleans columns."""
        try:
            df = yf.download(ticker, start=start_date, end=end_date, interval=timeframe, progress=False)
            
            if df.empty:
                print(f"âš ï¸ Error: No data for {ticker}. Check dates/ticker.") # âš ï¸ log message
                return None
            
            df = df.reset_index() 
            
            if isinstance(df.columns, pd.MultiIndex):
                df.columns = [col[0] for col in df.columns.values]

            new_columns = [col.lower() if isinstance(col, str) else col for col in df.columns]
            df.columns = new_columns 
            
            if 'date' not in df.columns:
                df.rename(columns={'index': 'date'}, inplace=True)
            df['date'] = pd.to_datetime(df['date'])

            print(f"ğŸ§¹ Final DataFrame columns: {df.columns.tolist()}") # ğŸ§¹ log message
            print("âœ… Data downloaded successfully (as a DataFrame).") # âœ… log message
            return df
        except Exception as e:
            error_line(f"Download or Data Processing error for {ticker}: {e}")
            return None

    
     # --- FINALIZED: draw_data method (Saves ALL images, Trades on THRESHOLD) ---
    def draw_data(self, df: pd.DataFrame, images_folder: str, symbol: str, tf: str):
        """
        Generates charts for every window, makes predictions, saves ALL images (for video),
        but only executes and records trades when the confidence threshold is met.
        """  
        try:
            # ğŸ“ Charting Constants
            height = 800  
            width = 1000  
            bars = self.bars # Dynamic bar count
            DISPLAY_FREQUENCY = 50 
            
            # ğŸ”‘ TRADING THRESHOLD
            CONFIDENCE_THRESHOLD = self.threshold # Dynamic threshold
            
            if len(df) < bars + 1:
                print(f"âš ï¸ Not enough data ({len(df)} rows) for a {bars}-bar window + outcome. Stopping.") # âš ï¸ log message
                return None, 0, 0

            prediction = []
            prediction_close_price=[]
            actual_outcome =[]
            create_folder(images_folder)
            
            chart_count = 0
            trade_executed_count = 0 
            
            # Delete old image files before starting (clean up)
            files = glob.glob(f'{images_folder}/*.jpg')
            for f in files:
                os.remove(f)
            
            # ğŸ”„ Start of the main iteration loop
            for start_index in range(len(df) - bars - 1): 
                chart_count += 1
                graph = np.zeros((height, width, 3), dtype=np.uint8)
                graph.fill(255) # White background
                
                # --- CANDLE DRAWING LOGIC (Using the current window) ---
                df_window = df.iloc[start_index : start_index + bars].copy()
                window_min = df_window[['low', 'close']].min().min()
                window_max = df_window[['high', 'open']].max().max()
                padding = (window_max - window_min) * 0.05
                scale_min = window_min - padding
                scale_range = (window_max + padding) - scale_min
                if scale_range == 0: scale_range = 1.0 

                candle_width_px = max(1, (width - 10) // bars // 2) 
                x_start = 5
                
                for i in range(bars): 
                    data_index = start_index + i 
                    original_data = df.iloc[data_index]

                    # Normalize and scale
                    norm_open = (original_data['open'] - scale_min) / scale_range
                    norm_close = (original_data['close'] - scale_min) / scale_range
                    norm_high = (original_data['high'] - scale_min) / scale_range
                    norm_low = (original_data['low'] - scale_min) / scale_range
                    
                    scale_factor = height - 20
                    y_open = int(height - 20 - (norm_open * scale_factor))
                    y_close = int(height - 20 - (norm_close * scale_factor))
                    y_high = int(height - 20 - (norm_high * scale_factor))
                    y_low = int(height - 20 - (norm_low * scale_factor))

                    # Colors
                    if original_data['open'] < original_data['close']:
                        body_color = (0, 255, 0) # Green ğŸŸ¢
                        body_top = y_close
                        body_bottom = y_open
                    else:
                        body_color = (0, 0, 255) # Red ğŸ”´
                        body_top = y_open
                        body_bottom = y_close

                    current_x = x_start + i * (2 * candle_width_px + 1)
                    wick_color = (0, 0, 0) # Black 
                    
                    # Draw thick wicks
                    if y_high < body_top:
                        cv2.rectangle(graph, (current_x - candle_width_px, y_high), 
                                              (current_x + candle_width_px, body_top), wick_color, -1) 
                    if y_low > body_bottom: 
                        cv2.rectangle(graph, (current_x - candle_width_px, body_bottom), 
                                              (current_x + candle_width_px, y_low), wick_color, -1)
                                              
                    # Draw candle body
                    cv2.rectangle(graph, (current_x - candle_width_px, body_top), 
                                    (current_x + candle_width_px, body_bottom), body_color, -1)
                # --- END CANDLE DRAWING LOGIC ---
                
                # ğŸ¤– STEP 1: Make Prediction
                results = self.model.predict(graph, conf=CONFIDENCE_THRESHOLD, verbose=False) 
                
                # --- ğŸ”‘ STEP 2: Confidence Check and Trade Preparation ---
                best_pred = None
                best_conf = 0.0
                
                # Check if any detection passed the confidence filter (i.e., best_pred is not None)
                if results and results[0].boxes and len(results[0].boxes) > 0:
                    box = results[0].boxes[0] 
                    best_conf = float(box.conf[0].item())
                    class_id = int(box.cls[0].item()) 
                    best_pred = self.model.names[class_id] # Valid trade signal

                # --- ğŸ–¼ï¸ STEP 3A: RENDER AND SAVE IMAGE (ALWAYS RUNS) ---
                # This renders the image, drawing bounding boxes ONLY for predictions that passed the threshold filter.
                render_np = render_result(model=self.model, image=graph, result=results[0])
                filename = f"graph_{chart_count:04d}.jpg" 
                output_path = f'./{images_folder}/{filename}'
                cv2.imwrite(output_path, render_np)
                # print(f"ğŸ’¾ Image {chart_count} saved to {output_path}.") # ğŸ’¾ log message
                
                # ğŸ“¢ Display a sample chart every 50 iterations
                if chart_count % DISPLAY_FREQUENCY == 0: 
                    print(f"--- ğŸ” **VISUAL CHECK:** Displaying Chart #{chart_count} ---") 
                    display(Image(filename=output_path, width=800)) 
                
                # --- ğŸ“ˆ STEP 3B: TRADE EXECUTION AND RECORD (ONLY IF THRESHOLD IS MET) ---
                if best_pred is not None:
                    print(f"âœ… Trade executed with {best_pred} (Conf: {best_conf:.2f})") # âœ… log message
                    trade_executed_count += 1 
                    
                    last_window_index = start_index + bars - 1  
                    outcome_index = start_index + bars  
                    
                    prediction.append(best_pred)
                    prediction_close_price.append(df['close'].iloc[last_window_index])
                    actual_outcome.append(df['close'].iloc[outcome_index])
                else:
                    print("â¸ï¸ No trade executed (Confidence below threshold or no detection)") # â¸ï¸ log message
            # ğŸ›‘ End of the main 'for' loop
            
            # -------------------------------------------------------------------
            # This final block MUST align with the 'for' loop to run only once after all charts are processed.
            # -------------------------------------------------------------------

            # ğŸ“ˆ Run Backtest & Calculate CAGR
            win_loss = backtest(prediction, prediction_close_price, actual_outcome) 
            
            total_trades = win_loss[1]
            cumulative_return_factor = win_loss[2]
            
            # Annualization (CAGR) Calculation
            cagr = 0.0
            if total_trades > 0: 
                start_date = df['date'].iloc[bars] 
                end_date = df['date'].iloc[-1]   
                total_days = (end_date - start_date).days
                total_years = total_days / 365.25 
                if total_years > 0:
                    cagr = (cumulative_return_factor ** (1 / total_years)) - 1
            
            print("\n> -----------------------------------")
            print(f"> ğŸ† **FINAL BACKTEST RESULTS** ({symbol} - {tf}) ğŸš€")
            print(f"> -----------------------------------")
            print(f">   ğŸ“ˆ **Win Rate:** {win_loss[0]:.2f}%")
            print(f">   ğŸ”„ **Total Trades:** {total_trades}")
            print(f">   ğŸ’° **Total Return:** {(cumulative_return_factor - 1.0) * 100.0:.2f}%")
            print(f">   ğŸ“ˆ **Annualized Returns (CAGR):** **{cagr * 100.0:.2f}%**")
            print("> -----------------------------------")
            
            print(f"ğŸ–¼ï¸ Saved **{chart_count} total charts** (Executed {trade_executed_count} trades) to /{images_folder}/") 
            
            # Return chart_count (total images saved) for video creation
            return images_folder, chart_count, total_trades 
                
        except Exception as e:
            print(f"âŒ Strategy execution error: {e}") # âŒ log message
            return None, 0, 0


    
    # --- ADDED: create_video_and_display method (CORRECTLY INDENTED) ---
    def create_video_and_display(self, images_folder: str, total_charts: int, output_filename: str = 'strategy_backtest.mp4'):
        """Stitches images into an MP4 video and displays it."""
        if total_charts == 0:
            print("âš ï¸ Skipping video creation: No charts were generated.") # âš ï¸ log message
            return
            
        image_files = sorted(glob.glob(f'{images_folder}/graph_*.jpg'))
        
        if not image_files:
            print("âš ï¸ Skipping video creation: No image files found.") # âš ï¸ log message
            return

        first_frame = cv2.imread(image_files[0])
        height, width, layers = first_frame.shape
        size = (width, height)
        fps = 10 

        fourcc = cv2.VideoWriter_fourcc(*'VP90') 
        try:
            out = cv2.VideoWriter(output_filename, fourcc, fps, size)
        except Exception:
            fourcc = cv2.VideoWriter_fourcc(*'mp4v')
            out = cv2.VideoWriter(output_filename, fourcc, fps, size)

        print(f"ğŸ¬ Starting video compilation of {len(image_files)} frames...") # ğŸ¬ log message

        for filename in image_files:
            img = cv2.imread(filename)
            out.write(img)
            
        out.release()
        print(f"âœ… Video compiled successfully: {output_filename}") # âœ… log message

        try:
            print("ğŸ’» Attempting robust video display...") # ğŸ’» log message
            
            with open(output_filename, "rb") as f:
                encoded_video = base64.b64encode(f.read()).decode()
            
            html = f"""
            <video width="{width}" height="{height}" controls autoplay loop>
                <source src="data:video/mp4;base64,{encoded_video}" type="video/mp4">
                Your browser does not support the video tag.
            </video>
            """
            display(HTML(html))
            print("âœ… Video display attempt completed.") # âœ… log message
            
        except Exception as e:
            error_line(f"Critical error during video display: {e}. Please download and view the output file.")
    # --- END ADDED: create_video_and_display method ---
```

```python
class TradingBot:
    """A trading class that handles data download, chart generation, prediction, and video output."""
    
    def __init__(self, model_path: str, bars: int = 40, threshold: float = 0.6):
        self.model = load_model(model_path)
        if self.model is None:
            raise Exception("Model could not be loaded. Cannot proceed.")
        self.bars = bars 
        self.threshold = threshold 
        print(f"âš™ï¸ Bot initialized with {self.bars} bars and a trade confidence threshold of {self.threshold}.") # âš™ï¸ log message

    # --- 1. DOWNLOAD DATA ---
    def download_data(self, ticker: str, start_date: str, end_date: str, timeframe: str = '1d') -> pd.DataFrame or None:
        """Downloads stock data and cleans columns."""
        try:
            df = yf.download(ticker, start=start_date, end=end_date, interval=timeframe, progress=False)
            
            if df.empty:
                print(f"âš ï¸ Error: No data for {ticker}. Check dates/ticker.") # âš ï¸ log message
                return None
            
            df = df.reset_index() 
            
            if isinstance(df.columns, pd.MultiIndex):
                df.columns = [col[0] for col in df.columns.values]

            new_columns = [col.lower() if isinstance(col, str) else col for col in df.columns]
            df.columns = new_columns 
            
            if 'date' not in df.columns:
                df.rename(columns={'index': 'date'}, inplace=True)
            df['date'] = pd.to_datetime(df['date'])

            print(f"ğŸ§¹ Final DataFrame columns: {df.columns.tolist()}") # ğŸ§¹ log message
            print("âœ… Data downloaded successfully (as a DataFrame).") # âœ… log message
            return df
        except Exception as e:
            error_line(f"Download or Data Processing error for {ticker}: {e}")
            return None

    # --- 2. DRAW CHARTS, PREDICT, AND BACKTEST ---
    def draw_data(self, df: pd.DataFrame, images_folder: str, symbol: str, tf: str):
        """
        Generates charts for every window, makes predictions, saves ALL images (for video),
        but only executes and records trades when the confidence threshold is met.
        """  
        try:
            # ğŸ“ Charting Constants
            height = 800  
            width = 1000  
            bars = self.bars 
            DISPLAY_FREQUENCY = 50 
            
            # ğŸ”‘ TRADING THRESHOLD
            CONFIDENCE_THRESHOLD = self.threshold 
            
            if len(df) < bars + 1:
                print(f"âš ï¸ Not enough data ({len(df)} rows) for a {bars}-bar window + outcome. Stopping.") # âš ï¸ log message
                return None, 0, 0, []
            
            # --- Initialization ---
            prediction = []
            prediction_close_price=[]
            actual_outcome =[]
            trade_image_files = [] # ğŸŒŸ List to hold paths of images where a trade occurred
            create_folder(images_folder)
            
            chart_count = 0
            trade_executed_count = 0 
            
            # Delete old image files...
            files = glob.glob(f'{images_folder}/*.jpg')
            for f in files:
                os.remove(f)
            
            # ğŸ”„ Start of the main iteration loop
            for start_index in range(len(df) - bars - 1): 
                chart_count += 1
                graph = np.zeros((height, width, 3), dtype=np.uint8)
                graph.fill(255) # White background
                
                # --- CANDLE DRAWING LOGIC ---
                df_window = df.iloc[start_index : start_index + bars].copy()
                window_min = df_window[['low', 'close']].min().min()
                window_max = df_window[['high', 'open']].max().max()
                padding = (window_max - window_min) * 0.05
                scale_min = window_min - padding
                scale_range = (window_max + padding) - scale_min
                if scale_range == 0: scale_range = 1.0 

                candle_width_px = max(1, (width - 10) // bars // 2) 
                x_start = 5
                
                for i in range(bars): 
                    data_index = start_index + i 
                    original_data = df.iloc[data_index]

                    # Normalize and scale
                    norm_open = (original_data['open'] - scale_min) / scale_range
                    norm_close = (original_data['close'] - scale_min) / scale_range
                    norm_high = (original_data['high'] - scale_min) / scale_range
                    norm_low = (original_data['low'] - scale_min) / scale_range
                    
                    scale_factor = height - 20
                    y_open = int(height - 20 - (norm_open * scale_factor))
                    y_close = int(height - 20 - (norm_close * scale_factor))
                    y_high = int(height - 20 - (norm_high * scale_factor))
                    y_low = int(height - 20 - (norm_low * scale_factor))

                    # Colors
                    if original_data['open'] < original_data['close']:
                        body_color = (0, 255, 0) # Green ğŸŸ¢
                        body_top = y_close
                        body_bottom = y_open
                    else:
                        body_color = (0, 0, 255) # Red ğŸ”´
                        body_top = y_open
                        body_bottom = y_close

                    current_x = x_start + i * (2 * candle_width_px + 1)
                    wick_color = (0, 0, 0) # Black 
                    
                    # Draw thick wicks
                    if y_high < body_top:
                        cv2.rectangle(graph, (current_x - candle_width_px, y_high), 
                                              (current_x + candle_width_px, body_top), wick_color, -1) 
                    if y_low > body_bottom: 
                        cv2.rectangle(graph, (current_x - candle_width_px, body_bottom), 
                                              (current_x + candle_width_px, y_low), wick_color, -1)
                                              
                    # Draw candle body
                    cv2.rectangle(graph, (current_x - candle_width_px, body_top), 
                                    (current_x + candle_width_px, body_bottom), body_color, -1)
                # --- END CANDLE DRAWING LOGIC ---
                
                # ğŸ¤– STEP 1: Make Prediction
                results = self.model.predict(graph, conf=CONFIDENCE_THRESHOLD, verbose=False) 
                
                # --- ğŸ”‘ STEP 2: Confidence Check and Trade Preparation ---
                best_pred = None
                best_conf = 0.0
                
                if results and results[0].boxes and len(results[0].boxes) > 0:
                    box = results[0].boxes[0] 
                    best_conf = float(box.conf[0].item())
                    class_id = int(box.cls[0].item()) 
                    best_pred = self.model.names[class_id] # Valid trade signal

                # --- ğŸ–¼ï¸ STEP 3A: RENDER AND SAVE IMAGE (ALWAYS RUNS) ---
                render_np = render_result(model=self.model, image=graph, result=results[0])
                filename = f"graph_{chart_count:04d}.jpg" 
                output_path = f'./{images_folder}/{filename}'
                cv2.imwrite(output_path, render_np)
                # print(f"ğŸ’¾ Image {chart_count} saved to {output_path}.") # ğŸ’¾ log message
                
                # ğŸ“¢ Display a sample chart every 50 iterations
                # if chart_count % DISPLAY_FREQUENCY == 0: 
                    # print(f"--- ğŸ” **VISUAL CHECK:** Displaying Chart #{chart_count} ---") 
                    # display(Image(filename=output_path, width=800)) 
                
                # --- ğŸ“ˆ STEP 3B: TRADE EXECUTION AND RECORD ---
                if best_pred is not None:
                    print(f"âœ… Trade executed with {best_pred} (Conf: {best_conf:.2f})") # âœ… log message
                    trade_executed_count += 1 
                    
                    trade_image_files.append(output_path) # ğŸŒŸ Record the path for random display
                    
                    last_window_index = start_index + bars - 1  
                    outcome_index = start_index + bars  
                    
                    prediction.append(best_pred)
                    prediction_close_price.append(df['close'].iloc[last_window_index])
                    actual_outcome.append(df['close'].iloc[outcome_index])
                # else:
                    # print("â¸ï¸ No trade executed (Confidence below threshold or no detection)") # â¸ï¸ log message
            # ğŸ›‘ End of the main 'for' loop
            
            # -------------------------------------------------------------------
            # This final block runs only once after the 'for' loop finishes.
            # -------------------------------------------------------------------

            # ğŸ“ˆ Run Backtest & Calculate CAGR
            win_loss = backtest(prediction, prediction_close_price, actual_outcome) 
            
            total_trades = win_loss[1]
            cumulative_return_factor = win_loss[2]
            
            # Annualization (CAGR) Calculation
            cagr = 0.0
            if total_trades > 0: 
                start_date = df['date'].iloc[bars] 
                end_date = df['date'].iloc[-1]   
                total_days = (end_date - start_date).days
                total_years = total_days / 365.25 
                if total_years > 0:
                    cagr = (cumulative_return_factor ** (1 / total_years)) - 1
            
            print("\n> -----------------------------------")
            print(f"> ğŸ† **FINAL BACKTEST RESULTS** ({symbol} - {tf}) ğŸš€")
            print(f"> -----------------------------------")
            print(f">   ğŸ“ˆ **Win Rate:** {win_loss[0]:.2f}%")
            print(f">   ğŸ”„ **Total Trades:** {total_trades}")
            print(f">   ğŸ’° **Total Return:** {(cumulative_return_factor - 1.0) * 100.0:.2f}%")
            print(f">   ğŸ“ˆ **Annualized Returns (CAGR):** **{cagr * 100.0:.2f}%**")
            print("> -----------------------------------")
            
            print(f"ğŸ–¼ï¸ Saved **{chart_count} total charts** (Executed {trade_executed_count} trades) to /{images_folder}/") 
            
            # Return chart_count (total images saved) for video creation
            return images_folder, chart_count, total_trades, trade_image_files,cagr
                
        except Exception as e:
            print(f"âŒ Strategy execution error: {e}") # âŒ log message
            return None, 0, 0, []

    # --- 3. CREATE VIDEO AND DISPLAY ---
    def create_video_and_display(self, images_folder: str, total_charts: int, output_filename: str = 'strategy_backtest.mp4'):
        """Stitches images into an MP4 video and displays it."""
        if total_charts == 0:
            print("âš ï¸ Skipping video creation: No charts were generated.") # âš ï¸ log message
            return
            
        image_files = sorted(glob.glob(f'{images_folder}/graph_*.jpg'))
        
        if not image_files:
            print("âš ï¸ Skipping video creation: No image files found.") # âš ï¸ log message
            return

        first_frame = cv2.imread(image_files[0])
        height, width, layers = first_frame.shape
        size = (width, height)
        fps = 10 

        # Attempt VP90 (better quality/compression) then fallback to mp4v
        fourcc = cv2.VideoWriter_fourcc(*'VP90') 
        try:
            out = cv2.VideoWriter(output_filename, fourcc, fps, size)
        except Exception:
            fourcc = cv2.VideoWriter_fourcc(*'mp4v')
            out = cv2.VideoWriter(output_filename, fourcc, fps, size)

        print(f"ğŸ¬ Starting video compilation of {len(image_files)} frames...") # ğŸ¬ log message

        for filename in image_files:
            img = cv2.imread(filename)
            out.write(img)
            
        out.release()
        print(f"âœ… Video compiled successfully: {output_filename}") # âœ… log message

        try:
            print("ğŸ’» Attempting robust video display...") # ğŸ’» log message
            
            with open(output_filename, "rb") as f:
                encoded_video = base64.b64encode(f.read()).decode()
            
            html = f"""
            <video width="{width}" height="{height}" controls autoplay loop>
                <source src="data:video/mp4;base64,{encoded_video}" type="video/mp4">
                Your browser does not support the video tag.
            </video>
            """
            display(HTML(html))
            print("âœ… Video display attempt completed.") # âœ… log message
            
        except Exception as e:
            error_line(f"Critical error during video display: {e}. Please download and view the output file.")

    # --- 4. DISPLAY RANDOM TRADE IMAGES ---
    def display_random_trade_images(self, trade_image_files: list, num_samples: int = 12):
        """Selects a random sample of trade images and displays them."""
        
        if not trade_image_files:
            print("ğŸš« No trade images were recorded to display.") # ğŸš« log message
            return

        sample_size = min(num_samples, len(trade_image_files))
        
        # Randomly select the image file paths
        selected_files = random.sample(trade_image_files, sample_size)
        
        print(f"\nğŸ–¼ï¸ **Displaying {sample_size} Random Trade Samples:**") # ğŸ–¼ï¸ log message
        
        # Prepare HTML to display images in a grid (e.g., 4 columns)
        images_per_row = 4
        html_content = ""
        
        for i, file_path in enumerate(selected_files):
            # Start a new row every 'images_per_row' images
            if i % images_per_row == 0:
                if i != 0:
                    html_content += "</div>" # Close previous row
                html_content += '<div style="display: flex; justify-content: space-around; margin-bottom: 20px;">'
            
            # Image display block
            html_content += f"""
            <div style="flex: 1; max-width: 250px; padding: 5px; text-align: center;">
                <p style="margin: 0; font-size: 10px; color: #555;">{file_path.split('/')[-1]}</p>
                <img src="{file_path}" style="width: 100%; height: auto; border: 1px solid #ccc;">
            </div>
            """
        
        # Close the final row div
        if html_content:
            html_content += "</div>"
            
        display(HTML(html_content))
```

```python
# --- DRIVER CODE AND EXECUTION ---

# ğŸ”— Using the model path provided for the Kaggle environment
MODEL_PATH = "/kaggle/input/futurepredictions/pytorch/default/1/best.pt" 

# Note: The TradingBot is instantiated inside the pipeline function.

def run_backtest_pipeline(
    ticker: str, 
    start_date: str, 
    end_date: str, 
    bars: int,              # ğŸ› ï¸ New required parameter
    threshold: float,       # ğŸ› ï¸ New required parameter
    timeframe: str = "1d"
):
    """
    Runs the full backtesting pipeline: data download, chart generation,
    prediction, backtest, video compilation/display, and random trade image display.
    
    The TradingBot instance is created on each run with the specified
    bars and threshold parameters.
    """
    
    # ğŸ¤– 1. Initialize the Bot instance for this specific run
    try:
        print(f"ğŸ§  Initializing Bot for run with **{bars} bars** and **{threshold} threshold**...") # ğŸ§  log message
        # ğŸ’¡ Bot is instantiated here with the run-specific parameters
        bot_instance = TradingBot(MODEL_PATH, bars=bars, threshold=threshold)
    except Exception as e:
        error_line(f"TradingBot Initialization Failed for {ticker}: {e}")
        return

    # ğŸ“Š Dynamic Trading Parameters
    TICKER_SYMBOL = ticker
    TIMEFRAME = timeframe
    START = start_date
    END = end_date
    OUTPUT_FOLDER = f"charts_{TICKER_SYMBOL}_{TIMEFRAME}_yolo_preds_{bars}b_{threshold:.2f}t"
    VIDEO_FILENAME = f"{TICKER_SYMBOL}_{TIMEFRAME}_backtest_{bars}b_{threshold:.2f}t.mp4"

    try:
        print(f"\n--- ğŸš€ Starting Backtest for **{TICKER_SYMBOL}** ({START} to {END}) ---") # ğŸš€ log message
        
        # ğŸ“¥ 2. Download Data
        stock_df = bot_instance.download_data(TICKER_SYMBOL, START, END, TIMEFRAME)
        
        if stock_df is not None:
            # ğŸ–¼ï¸ 3. Draw Charts, Predict, and Backtest
            # ğŸŒŸ UPDATED: Capturing the new 'trade_files' list
            images_folder, chart_count, total_trades, trade_files , final_cagr = bot_instance.draw_data(
                stock_df, OUTPUT_FOLDER, TICKER_SYMBOL, TIMEFRAME
            )
            
            # # ğŸ¬ 4. Create and Display Video
            # if images_folder:
            #     bot_instance.create_video_and_display(images_folder, chart_count, VIDEO_FILENAME)
            
            # ğŸ–¼ï¸ 5. Display Random Trade Images
            if trade_files:
                bot_instance.display_random_trade_images(trade_files, num_samples=12)
        
        print(f"--- âœ… Pipeline Complete for **{TICKER_SYMBOL}** ---") # âœ… log message
        return final_cagr

    except Exception as e:
        error_line(f"Pipeline execution failed for {TICKER_SYMBOL}: {e}")
```

```python
import optuna 


# ğŸ”— Using the model path provided for the Kaggle environment
MODEL_PATH = "/kaggle/input/futurepredictions/pytorch/default/1/best.pt" 

# --- OPTUNA OBJECTIVE FUNCTION ---

def objective(trial: optuna.Trial):
    """
    Optuna objective function to maximize CAGR by tuning the 'bars' parameter.
    
    NOTE: It's common practice to disable video/image saving inside the objective 
    to maximize optimization speed.
    """
    
    # âš™ï¸ 1. Suggest a parameter value for 'bars' (the lookback window)
    # The 'bars' parameter must be an integer and large enough for charting.
    # We suggest a range common for candlestick analysis, e.g., 20 to 100.
    bars_to_test = trial.suggest_int('bars', low=20, high=100, step=5)
    
    # 2. Define fixed parameters for the optimization run
    # These values are fixed while we optimize 'bars'.
    FIXED_TICKER = "AAPL"
    FIXED_START_DATE = "2014-01-01"
    FIXED_END_DATE = "2025-01-01"
    FIXED_THRESHOLD = 0.70 # Keeping the threshold fixed for now
    
    print(f"ğŸ¯ Trial {trial.number}: Testing bars={bars_to_test}") # ğŸ¯ log message
    
    # 3. Execute the backtest pipeline with the suggested parameter
    # The pipeline is expected to return the final_cagr (Annualized Return).
    final_cagr = run_backtest_pipeline(
        ticker=FIXED_TICKER, 
        start_date=FIXED_START_DATE, 
        end_date=FIXED_END_DATE, 
        bars=bars_to_test, 
        threshold=FIXED_THRESHOLD,
        timeframe="1d"
    )
    
    # Optuna maximizes the return value, so we return the CAGR
    # Since CAGR is a percentage, we convert it to a float (e.g., 0.15) 
    # and return it. The pipeline returns the raw float (e.g., 0.15).
    return final_cagr


# --- OPTUNA DRIVER FUNCTION ---

def run_optimization(n_trials: int = 50, study_name: str = 'TradingBot_Bars_Optimization'):
    """
    Creates and runs the Optuna study.
    """
    
    # Create the study object. We want to MAXIMIZE the return (CAGR).
    study = optuna.create_study(
        direction='maximize',
        study_name=study_name,
        load_if_exists=True # Useful to continue previous studies
    )
    
    print(f"\n--- ğŸ§ª Starting Optuna Study: {study_name} ({n_trials} trials) ---") # ğŸ§ª log message
    
    # Run the optimization
    study.optimize(objective, n_trials=n_trials, show_progress_bar=True)
    
    # Output the results
    print("\n--- ğŸ† Optimization Complete ---")
    print(f"â­ Best trial number: {study.best_trial.number}")
    print(f"ğŸ’° Best CAGR (ARR): {study.best_value * 100.0:.2f}%")
    print(f"ğŸ“ˆ Best parameters: {study.best_params}")

    # Optional: Display optimization plot after execution
    # import optuna.visualization as viz
    # viz.plot_optimization_history(study).show()
    
    return study.best_params
```

```python
import os
import zipfile
from os.path import join, isfile

def zip_all_zips_into_one(base_dir='/kaggle/working', output_zip_name='master_archive.zip'):
    # ğŸ¯ Define the full path for the master output zip
    output_zip_path = join(base_dir, output_zip_name)
    
    # ğŸ“ List all files in the base directory
    all_items = os.listdir(base_dir)
    
    # ğŸ” Filter for only the existing ZIP files, excluding the new master archive
    zip_files_to_compress = [
        item for item in all_items 
        if item.endswith('.zip') and isfile(join(base_dir, item)) and item != output_zip_name
    ]

    if not zip_files_to_compress:
        print(f"ğŸ’¡ No .zip files found in {base_dir} to consolidate (excluding '{output_zip_name}').")
        return

    # ğŸš€ Create the master ZIP file
    print(f"ğŸ“¦ Creating master archive: {output_zip_name}")
    try:
        with zipfile.ZipFile(output_zip_path, 'w', zipfile.ZIP_DEFLATED) as master_zip:
            for zip_file_name in zip_files_to_compress:
                full_path = join(base_dir, zip_file_name)
                
                # The second argument (arcname) ensures the file name inside the zip is clean
                master_zip.write(full_path, arcname=zip_file_name)
                
                # ğŸ¥³ Log the file being added
                # print(f"  -> Added {zip_file_name}") 
        
        print(f"âœ… Master archive '{output_zip_name}' successfully created containing {len(zip_files_to_compress)} files.")

    except Exception as e:
        print(f"ğŸ”¥ An error occurred during zipping: {e}")

# Example Usage:
print("--- Starting ZIP Consolidation ---")
zip_all_zips_into_one()
print("--- Consolidation Complete ---")
```
